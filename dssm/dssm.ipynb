```python
rugpt_generator = pipeline("text-generation", model="sberbank-ai/rugpt3large_based_on_gpt2")
def get_gpt_answer(prompt):
    result = rugpt_generator(prompt, max_new_tokens=50, do_sample=True, temperature=0.7)
    return result[0]["generated_text"][len(prompt):]
```

```python
def generate_positive_query(book_name, description):
    #prompt = "Придумай точный поисковый запрос из одного слова для Google, чтобы найти книгу по её описанию.\nНазвание книги: " + book_name + "\nУчитывай описание книги:" + description + "\nПоисковый запрос: \""
    prompt = f"""
Ты — экспертный алгоритм по подбору книг. На основе описания книги сгенерируй точный поисковый запрос,
который поможет найти эту книгу в интернете или магазине. **Не используй название книги** — только жанр, автора,
тематику, уникальные элементы сюжета и другие ключевые детали из описания.

Правила:
1. Запрос должен быть конкретным, но без названия.
2. Используй только информацию из данного описания.
3. Учитывай возможные синонимы и альтернативные формулировки темы.
Название книги:
{book_name}

Описание книги:
{description}
Искомый запрос: \" """
    res = ""
    while True:
        res = get_gpt_answer(prompt)
        #print(res)
        ind = res.find("\"")
        if ind == -1 or ind > 40:
            continue
        res = res[:ind].strip()
        #print(res)
        if res.lower() == book_name.lower():
            continue
        break
    return res
```

```python
def generate_negative_query(book_name, description):
    prompt = f"""Ты — опытный библиотекарь, который разбирается в литературе и поисковых запросах.
Твоя задача — придумать такой поисковый запрос (например, для Google или книжного магазина),
чтобы книга с указанным названием и описанием НЕ попадала в результаты поиска по этому запросу.
Проанализируй название и описание книги."\
Придумай запрос, который: Лексически или тематически связан с книгой, но не соответствует её содержанию;
Может включать слова из названия, но в другом контексте; Может отсылать к жанру, но с искажением ключевой темы.
Название книги:
{book_name}

Описание книги:
{description}
Поисковый запрос, исключающий книгу: \""""
    res = ""
    while True:
        res = get_gpt_answer(prompt)
        #print(res)
        ind = res.find("\"")
        if ind == -1 or ind > 40:
            continue
        res = res[:ind].strip()
        #print(res)
        if res.lower() == book_name.lower():
            continue
        break
    return res
```

```python
bert_emb = 384
```

```python
class BertDataset(Dataset):
    def __init__(self, queries, docs, labels, bert_model):
        self.queries = bert_model.encode(queries, convert_to_tensor=True)
        self.docs = bert_model.encode(docs, convert_to_tensor=True)
        self.labels = torch.FloatTensor(labels)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.queries[idx], self.docs[idx], self.labels[idx]
```

```python
train_dataset = BertDataset(queries, docs, labels, bert_model)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
def save_str_list(strings, path):
    strings = [s.replace('\n', '').replace('\r', '') for s in strings]

    with open(path, 'w', encoding='utf-8') as f:
        for line in strings:
            f.write(line + '\n')
```

```python
def save_int_list(ints, path):
    with open(path, 'w', encoding='utf-8') as f:
        for x in ints:
            f.write(str(x) + '\n')
```

```python
save_str_list(queries, "queries.txt")
save_str_list(docs, "docs.txt")
save_int_list(labels, "labels.txt")
```

#DSSM

```python
from tqdm.auto import tqdm
from IPython.display import clear_output

def train(num_epochs, model, criterion, opt, train_dl):
    model.train()
    for epoch in range(num_epochs):
        #print("start epoch", epoch)
        losses = []
        for it, batch in enumerate(tqdm(train_dl, desc='Training', leave=False)):
            queries, docs, labels = batch
            #print("it =", it)

            output = model(queries.to(device), docs.to(device))
            #print(output)
            opt.zero_grad()
            #print(output)
            loss = criterion(output, labels.to(device))
            losses.append(loss.item())
            #print(output)
            loss.backward()
            opt.step()

        print(f"Epoch {epoch}, Loss: {np.mean(np.array(losses))}")
```

```python


class DSSM(nn.Module):
    def __init__(self, hidden_dim=256):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(bert_emb, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
        )


    def forward(self, query, doc):
        query_embed = self.fc(query)
        doc_embed = self.fc(doc)

        return torch.cosine_similarity(query_embed, doc_embed, dim=1)
```

```python
model = DSSM().to(device)
num_epochs = 50
opt = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.BCEWithLogitsLoss()

train(num_epochs, model, criterion, opt, train_loader)
```

```python
num_epochs = 50
opt = torch.optim.Adam(model.parameters(), lr=1e-4)
criterion = nn.BCEWithLogitsLoss()

train(num_epochs, model, criterion, opt, train_loader)
```

```python
def search(query, top_k=10):
    query_emb = bert_model.encode(query, convert_to_tensor=True).to(device)
    res = []
    for i in range(len(all_docs)):
        doc_emb = bert_model.encode(all_docs[i], convert_to_tensor=True).to(device)
        dist = model(doc_emb.view(1, -1), query_emb.view(1, -1)).cpu()
        res.append((dist, i))
    #print(res)
    res = sorted(res)[-top_k:]

    res = list(reversed(res))
    ans = []
    for (d, i) in res:
        ans.append(all_docs[i])
    return ans
```
